```python
#Mol block display from mol 
#mofrom rdkit import Chem
from rdkit import Chem
from rdkit.Chem import AllChem

smiles = "C1=CC=CC=C1"
mol = Chem.MolFromSmiles(smiles)
 
#create molblock 
print(Chem.MolToMOlBlock(mol))

#generate 2d coords for mol then reate molblock 
AllChem.Compute2DCoords(mol)
molblock_2d = Chem.MolToMolBlock(mol)
print(molblock_2d)

#generate 3d coords for mol then create molblock, follow the seqence: 
#Add/remove hydrogens (if necessary).
#Embed the molecule to generate 3D coordinates.
#Optimize the 3D coordinates using a force field (MMFF).
mol_3d = Chem.Add/RemoveHs(mol)
params = AllChem.ETKDGv3()
params.randomSeed = 0xfood  #for reproducibility

AllChem.EmbedMolecule(mol_3d, params)
AllChem.MMFFOptimizeMolecule(mol_3d, maxIters=500) #use forcefield for optimization 
print(Chem.MolToMlBlock(mol_3d))
```
